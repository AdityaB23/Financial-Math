# Full ML Stat-Arb Pipeline

```
from src.data import load_prices, compute_returns
from src.features import create_features
from src.models import train_model
from src.walkforward import walk_forward
from src.backtest import backtest

prices = load_prices()
returns = compute_returns(prices)
features = create_features(returns)
```

```
# Target: next-day SPY return
y = returns['SPY'].shift(-1).dropna()
X = features.loc[y.index]
```

```
preds = walk_forward(X, y, lambda X,y: train_model(X,y,'ridge'))
```

```
strat, equity, sharpe, dd = backtest(preds, returns['SPY'])
print('Sharpe:', sharpe)
print('Max Drawdown:', dd)
```

```
import matplotlib.pyplot as plt
equity.plot(title='Equity Curve')
plt.show()
```
